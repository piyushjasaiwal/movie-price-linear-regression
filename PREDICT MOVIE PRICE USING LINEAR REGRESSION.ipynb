```python
import pandas
from pandas import DataFrame as df
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
#how to read data from a csv file using pandas dataframe

data = pandas.read_csv('cost_revenue_dirty.csv')
```

```python
#to know all the features from the data frame imported from csv file

data.describe()
```

```python
data.tail()
```

```python
type(data)
```

```python
x = df(data, columns = ["Production_Budget_USD"])
y = df(data = data, columns = ["Worldwide_Gross_USD"])
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x,y, alpha = 0.3,)
plt.xlim(0,425000000)
plt.ylim(0,2783918982)
plt.title('Film production cost vs Gross Revenue (in USD)',fontsize = 20)
plt.xlabel('Film Budget $' ,fontsize = 16)
plt.ylabel('Film Revenue $' ,fontsize = 16)
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(x,y)
plt.xlim(0,425000000)
plt.ylim(0,2783918982)
plt.title('Film production cost vs Gross Revenue (in USD)',fontsize = 20)
plt.xlabel('Film Budget $' ,fontsize = 16)
plt.ylabel('Film Revenue $' ,fontsize = 16)
plt.show()
```

```python
lr = LinearRegression()
lr.fit(x, y)
```

## Slope Coefficient

```python
lr.coef_       #thet 1 slope of the line
```

```python
#interept on the y axis

lr.intercept_
```

```python
#plotting the regression line

plt.figure(figsize = (10,6))
plt.scatter(x,y, alpha = 0.3)
plt.xlim(0,425000000)
plt.ylim(0,2783918982)
plt.title('Film production cost vs Gross Revenue (in USD)',fontsize = 20)
plt.xlabel('Film Budget $' ,fontsize = 16)
plt.ylabel('Film Revenue $' ,fontsize = 16)

plt.plot(x, lr.predict(x), color = '#A40606' ,linewidth = 1)

plt.show()
```

```python
#getting the r square value from the regression
lr.score(x, y)
```
